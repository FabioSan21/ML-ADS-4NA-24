```python
import pandas as pd

# URL do dataset Wine
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data'

# Nome das colunas em português BR
column_names_pt_br = [
    'classe',
    'alcool',
    'acido_malico',
    'cinzas',
    'alcalinidade_de_cinzas',
    'magnesio',
    'fenois_totais',
    'flavanoides',
    'fenois_nao_flavanoides',
    'proantocianinas',
    'intensidade_de_cor',
    'matiz',
    'od280_od315_de_vinhos_diluidos',
    'prolina'
]

# Ler o arquivo CSV com as colunas especificadas, definindo a coluna 'classe' como object
vinhos = pd.read_csv(url, names=column_names_pt_br, dtype={'classe': object})
```

```python
classe_counts = vinhos['classe'].value_counts()
estatisticas = vinhos[['alcool', 'acido_malico', 'magnesio', 'matiz']].describe()

print("Contagem de classes:")
print(classe_counts)
print("\nEstatísticas descritivas:")
print(estatisticas)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Configuração do estilo dos gráficos
sns.set(style="whitegrid")

# Colunas selecionadas para boxplots
colunas_boxplot = ['magnesio', 'prolina', 'cinzas', 'alcool']

# Criando a figura e os boxplots
plt.figure(figsize=(12, 8))
for i, coluna in enumerate(colunas_boxplot, start=1):
    plt.subplot(2, 2, i)
    sns.boxplot(data=vinhos, y=coluna, color='skyblue')
    plt.title(f'Boxplot de {coluna}')
    plt.ylabel(coluna)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt



# Configurando o estilo dos gráficos
plt.style.use('ggplot')

# Colunas selecionadas para boxplots segregados por classe
colunas_segregadas = ['magnesio', 'fenois_totais', 'acido_malico', 'alcool']

# Criando boxplots segregados por classe
plt.figure(figsize=(14, 10))
for i, coluna in enumerate(colunas_segregadas, start=1):
    plt.subplot(2, 2, i)
    vinhos.boxplot(column=coluna, by='classe', grid=False)
    plt.title(f'Boxplot de {coluna} por Classe')
    plt.suptitle("")  # Remover título geral automático
    plt.xlabel('Classe')
    plt.ylabel(coluna)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Criando o histograma para a coluna 'fenois_totais' com 10 bins
plt.figure(figsize=(8, 6))
sns.histplot(vinhos['fenois_totais'], bins=10, kde=False, color='blue')

# Configurando título e rótulos
plt.title('Histograma de Fenois Totais', fontsize=16)
plt.xlabel('Fenois Totais', fontsize=12)
plt.ylabel('Frequência', fontsize=12)

plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filtrando os dados para incluir apenas os vinhos da classe 3
vinhos_classe_3 = vinhos[vinhos['classe'] == '3']

# Criando o histograma para 'fenois_totais' apenas da classe 3 com 10 bins
plt.figure(figsize=(8, 6))
sns.histplot(vinhos_classe_3['fenois_totais'], bins=10, kde=False, color='green')

# Configurando título e rótulos
plt.title('Histograma de Fenois Totais - Classe 3', fontsize=16)
plt.xlabel('Fenois Totais', fontsize=12)
plt.ylabel('Frequência', fontsize=12)

plt.grid(True)
plt.show()
```
