## **Configuração do Ambiente**

```python
# Importar bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

## **Carregar e Visualizar os Dados**

```python
data = {
    'Nome_Planta': ['Planta A', 'Planta B', 'Planta C', 'Planta D', 'Planta E'],
    'Região': ['Norte', 'Sul', 'Sudeste', 'Nordeste', 'Centro-Oeste'],
    'Uso_Terapêutico': ['Dor de cabeça', 'Inflamação', 'Dor de cabeça', 'Febre', 'Gripe'],
    'Característica_1': [1, 0, 1, 0, 1],
    'Característica_2': [0, 1, 0, 1, 1]
}
df = pd.DataFrame(data)

df.head()
```

## **Análise Descritiva**

```python
# Informações gerais
df.info()

# Estatísticas descritivas
df.describe()
```

## **Visualização dos Dados**

```python
# Gráfico de barras por região
plt.figure(figsize=(10, 6))
sns.countplot(x='Região', data=df)
plt.title('Distribuição de Plantas Medicinais por Região')
plt.show()
```

## **Modelagem de Machine Learning**

```python
# Separar características e alvo
X = df.drop('Uso_Terapêutico', axis=1)
y = df['Uso_Terapêutico']

# Codificar variáveis categóricas em X
le_X = LabelEncoder()
for col in X.columns:
    if X[col].dtype == 'object':
        X[col] = le_X.fit_transform(X[col])

# Codificar a variável alvo
le_y = LabelEncoder()
y = le_y.fit_transform(y)

# Normalizar os dados
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Dividir dados em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)

# Treinar modelo Random Forest
rf = RandomForestClassifier()
rf.fit(X_train, y_train)

# Avaliar o modelo
y_pred = rf.predict(X_test)
print(classification_report(y_test, y_pred))
```

## **Insights e Conclusões**

- Identificamos quais características das plantas influenciam determinadas aplicações terapêuticas.
- As visualizações permitiram entender a distribuição regional das plantas medicinais.
- O modelo de machine learning ajudou a predizer possíveis usos terapêuticos com boa precisão.
